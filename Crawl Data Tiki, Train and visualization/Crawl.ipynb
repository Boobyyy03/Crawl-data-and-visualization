# Extract Crawl Data

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
from selenium.webdriver.common.keys import Keys
import pandas as pd

driver = webdriver.Chrome()
driver.get("https://tiki.vn/tieu-thuyet-phuong-tay/c67997")
sleep(5)

exist_link = []
current_window = driver.current_window_handle
product_name = []
price = []
quantity_sold = []
product = driver.find_elements(By.XPATH,"//a[@class='style__ProductLink-sc-139nb47-2 cKoUly product-item']")
for i in product:
    i.send_keys(Keys.CONTROL + Keys.RETURN)
    driver.switch_to.window(driver.window_handles[1])
    
    product_name_element = driver.find_elements(By.XPATH, '//h1[@class="Title__TitledStyled-sc-c64ni5-0 iXccQY"]')
    price_element = driver.find_elements(By.XPATH, '//div[@class="product-price__current-price"]')
    quantity_sold_element = driver.find_elements(By.XPATH, '//div[@data-view-id="pdp_quantity_sold"]')
    for name in product_name_element:
        product_name.append(name.text)
    for j in price_element:
        price.append(j.text)
    for k in quantity_sold_element:
        quantity_sold.append(k.text)
    sleep(2)
    driver.close()
    driver.switch_to.window(current_window)
data_frame = pd.DataFrame(list(zip(product_name,price,quantity_sold)),columns=['product_name','price','quantity_sold'])
driver.quit()
```

# Transform

```python
data_frame['price'] = data_frame['price'].str.replace('₫','').str.replace('.','')
```

```python
data_frame['quantity_sold'] = data_frame['quantity_sold'].str.split(' ').str[2].str.replace('+','')
```
